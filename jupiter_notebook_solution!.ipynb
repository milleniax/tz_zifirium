# <center>Анализ эффективности рекламы</center>

Иморт библиотек и запись файлов для анализа

```python
import pandas as pd
import numpy as np
```

```python
spend = pd.read_csv('./in_data_p.csv')
downloads = pd.read_csv('./in_data_a.csv')
```

<b>Анализ файлов</b>

```python
spend.rename(columns={'date': 'Date'}, inplace=True)
```

```python
spend.head(10)
```

```python
downloads = downloads.sort_values('Date')
```

```python
downloads.head(10)
```

```python
spend.describe()
```

```python
downloads.describe()
```

Две таблицы можно соединить по полям Date и ad_id

<b>Соединяем с помощью full join</b>

```python
all_data = spend.merge(downloads, on=['Date', 'ad_id'], how='outer')
```

```python
all_data
```

Появляется много пустот, которые уберутся при группировке

<b>Делаем итоговую таблицу out суммируя траты и установки</b>

```python
out = all_data.groupby(['Date', 'app', 'Campaign','os']).agg({'spend': 'sum', 'Installs': 'sum'})
```

<b>Создаем столбец, показывающий эффективность рекламы</b>

```python
out['cpi'] = out['spend']/out['Installs']
```

Заменяем пустые значения и бесконечности образовавшиеся вследствии деления на ноль(скачиваний)

```python
out['cpi'].replace([np.inf, np.NaN], 0, inplace=True)
```

```python
out
```

<b>Записываем в файл</b>

```python
out.to_csv('./out.csv', sep='\t')
```
